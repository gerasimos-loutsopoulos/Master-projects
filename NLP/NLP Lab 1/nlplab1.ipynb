```python
# Write a Python function that filters out the stopwords from 
# the BBC corpus
def filter_stopwords(corpus_text):
    # Tokenize the corpus text
    tokenized_corpus = word_tokenize(corpus_text, language='english')

    # Remove stopwords from the tokenized corpus
    stop_words = set(stopwords.words('english'))
    filtered_corpus = [word for word in tokenized_corpus if word not in stop_words]

    return filtered_corpus
```

```python
# 4) 
# How many tokens are there in the BBC corpus after filter out the stopwords
# are these reasonable or to many or too few?
tokens_after_filtering = filter_stopwords(corpusText)
print('length of filtered tokens',len(tokens_after_filtering))
```

```python
# Check if all stopwords really were removed from the BBC corpus
# Can you write a Python function that also matches initial uppercase stopwords
def filter_with_lowercase(corpus_text):
    tokenized_corpus = word_tokenize(corpus_text, language='english')
    stop_words = set(stopwords.words('english'))
    filtered_corpus = [word for word in tokenized_corpus if word.lower() not in stop_words]
    return filtered_corpus

print('check if all stopwords removed = ', len(tokens_after_filtering) == filter_with_lowercase(corpusText))
```

```python
# 5)
# How many tokens are there in the BBC corpus after filtering also
# in lower case?
print('number of tokens after removoing stopwords in lowercase = ', len(filter_with_lowercase(corpusText)))
```

```python
# Import the WordNet lemmatizer in NLTK
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
# Write a Python function that lemmatizes the tokens in the BBC corpus
def lemmitize(corpus_text):
    filtered_corpus = filter_with_lowercase(corpus_text)
    lemmatized_corpus = [lemmatizer.lemmatize(word) for word in filtered_corpus]

    return lemmatized_corpus
```

```python
# 7)
# How many types do you have in the BBC corpus after lemmatization?

print('numbre of types after lemmitization = ', len(set(lemmitize(corpusText))))
```

```python
# Write a Python function that stems the tokens in the BBC corpus


def stem(corpus_text):
    filtered_corpus = filter_with_lowercase(corpus_text)
    stemmed_corpus = [stemmer.stem(word) for word in filtered_corpus]
    return stemmed_corpus
```

10)
Count the number of Persons, Organisation and GPEs, Geo Political Entities in the BBC corpus. Follow the above example and extend the function.


The chunking of the BBC corpus takes some time, please be patient.

```python
print('Chunking')
```

```python
import nltk

# Download necessary NLTK resources
nltk.download('maxent_ne_chunker')
nltk.download('words')

# Your BBC corpus goes here

def count_entities(corpus_text):
    # Tokenization
    tokenized = nltk.word_tokenize(corpus_text)

    # Part-of-Speech Tagging
    tagged = nltk.pos_tag(tokenized)

    # Chunking (Named Entity Recognition)
    chunked = nltk.ne_chunk(tagged)

    # Initialize counts
    person_count = 0
    organization_count = 0
    gpe_count = 0

    # Iterate through named entities and count occurrences
    for chunk in chunked:
        if isinstance(chunk, nltk.Tree):
            entity_type = chunk.label()
            if entity_type == 'PERSON':
                person_count += 1
            elif entity_type == 'ORGANIZATION':
                organization_count += 1
            elif entity_type == 'GPE':
                gpe_count += 1

    return person_count, organization_count, gpe_count

# Call the function with your BBC corpus
person_count, organization_count, gpe_count = count_entities(corpusText)

# Print the results
print("Number of Persons:", person_count)
print("Number of Organizations:", organization_count)
print("Number of GPEs:", gpe_count)
```
